```python
tweets = []
tweet_count = 0
end_date = datetime.utcnow() - timedelta(days=365)
for status in tp.Cursor(api.user_timeline, id=target, tweet_mode='extended').items():
    if status.created_at < end_date:
        print("Completed Tweet Retrieval")
        break
    if status.full_text.startswith("@") == False and status.full_text.startswith("RT") == False:
        tweet = re.sub(r'https.*', '', status.full_text).replace("\n", " ").strip()
        tweet_count += 1
        tweets.append(tweet)
```

```python
tweetSeries = pd.Series(tweets).str.replace(r'\B(\@[ 0-9a-zA-Z_]+\b)(?!;)', "%%USER%%").str.replace(r'\s\s+', " ")
noHashtag = tweetSeries.str.replace(r'\B(\#[ 0-9a-zA-Z_]+\b)(?!;)', "%%HASHTAG%%").str.replace(r'\s\s+', " ")
noPunc = noHashtag.str.replace(r'[.,?!:;<>_\(\)\[\]\{\}\+\=\@\#\$\%\^\|\*]', "").str.replace(r'[-_‚Äî0-9]', " ")
sentList = noPunc.str.replace(" USER", "").str.replace(" HASHTAG", "").str.split()
tweetDF = tweetSeries.to_frame("Baseline")
tweetDF["Baseline + No HashTags"] = noHashtag
tweetDF["Baseline + No HashTags + No Punc"] = noPunc
tweetDF["List of Words"] = sentList
tweetDF["Part-of-Speech"] = tweetDF["List of Words"].apply(nltk.pos_tag)
tweetDF
#234
```

```python
nounsNadj = []
for wordlist in tweetDF["Part-of-Speech"]:
    nas = [word for word,pos in wordlist if pos.startswith("NN") or pos.startswith("JJ")] 
    if len(nas) >= 3:
        keywords = random.sample(nas, 3)
    elif len(nas) > 0:
        keywords = nas
    nounsNadj.append(keywords)

tweetDF["Keywords"] = nounsNadj
tweetDF
```

```python
#Word2Vec Implementation
sentences = tweetSeries.str.split().tolist()
model = gensim.models.Word2Vec(sentences, min_count=1)
print(model)
# summarize vocabulary
words = list(model.wv.vocab)
print(words)
print(model['collaborate'])
```
